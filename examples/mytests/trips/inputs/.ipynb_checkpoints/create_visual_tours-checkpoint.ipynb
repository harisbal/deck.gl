```python
import os
import datetime
import json
import pandas as pd
```

```python
def create_segment(trip):
    
    dt = trip['Departtime']
    tt = trip['Traveltime']
    
    srcx, srcy = trip[['Xsource', 'Ysource']]
    tgtx, tgty = trip[['Xtarget', 'Ytarget']]
       
    segments = [[srcx, srcy, dt], [tgtx, tgty, (dt+tt)]]
    return segments
```
